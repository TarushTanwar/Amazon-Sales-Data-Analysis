# **Now let's create heatmap to find correlation between features.**

```python
# Extract numeric columns
numeric_data = df.select_dtypes(include=['int64', 'float64'])
# Creating Heatmap
sns.heatmap(numeric_data.corr(method='pearson'),annot=True, cmap='Blues')
```

#Observations:

* As you can see in the above heatmap **Unit Price** and **Unit Cost** are Highly corelated.
* Also **Unit Price** moderately related to **Total Revenue** and **Total Cost** but less related to **Total Profit**.
* lastly their No relation between **Order ID** and rest other features.

```python
df['Item Type'].value_counts()
```

```python
wp = { 'linewidth' : 1, 'edgecolor' : "black" }
labels = ('Cloths','Cosmetics','Office Suppiles','Fruits','Personal Care','Household','Beverages','Baby Food','Cereal','Vegetables','Snacks','Meat')
colors = ('blue','cyan','blue','cyan','blue','cyan','blue','cyan','blue','cyan','blue','cyan')
plt.figure(figsize = (13, 6))
plt.pie(df['Item Type'].value_counts(),data = df,
        autopct = '%0.0f',shadow = False ,
        colors = colors,labels = labels,
        wedgeprops = wp)
plt.axis('equal')
plt.title('Which Item Covers Most Area')
plt.legend(df['Item Type'].value_counts(), fontsize= 13)
plt.show()
```

```python
numeric_data.groupby('Item Type').mean()['Units Sold'].plot(kind = 'bar',color = 'blue')
plt.title('Avg. Units sold by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Avg. Units Sold')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.groupby('Order Date').sum()['Total Revenue'].plot(kind = 'line',color = 'blue')
plt.title('Total Revenue over time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
print('*'*20,'Total Revenue:',df['Total Revenue'].sum(),'*'*23)
plt.show()
```

# Month-wise Sales

```python
numeric_data['Order Month'] = df['Order Date'].dt.month
numeric_data.groupby('Order Month').sum()['Total Revenue'].plot(kind = 'bar', color = 'blue')
plt.title('Total Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

# Year-wise Sales

```python
numeric_data['Order Year'] = df['Order Date'].dt.year
numeric_data.groupby('Order Year').sum()['Total Revenue'].plot(kind = 'bar', color = 'blue')
plt.title('Total Revenue By Year')
plt.xlabel('Order Year')
plt.ylabel('Total Revenue')
plt.show()
```

# Relation between Unit Price and Total Revenue.

As **Unit Price** increases **Total Revenue** Increases.

```python
plt.scatter(df['Unit Price'],df['Total Revenue'],color = 'blue')
plt.title('Unit Price Vs Total Revenue')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
```

```python
df.groupby('Region')['Total Revenue'].sum().plot(kind = 'bar', color = 'blue')
plt.title('Total Revenue By Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.show()
```

```python
sns.barplot(x= 'Order Priority', y= 'Total Profit',color = 'blue', data =df)
plt.show()
```
